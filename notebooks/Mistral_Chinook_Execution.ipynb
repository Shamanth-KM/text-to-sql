```python
##Convert Prediction file to json for Birdbench
import csv
import json

def convert_csv_to_custom_json(csv_file_path, json_file_path):
    output_data = {}

    with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for idx, row in enumerate(reader):
            clean_sql = row.get("clean_sql_extracted", "").strip()
            db_id = row.get("db_id", "").strip()
            combined = f"{clean_sql}\t----- bird -----\t{db_id}"
            output_data[str(idx)] = combined

    with open(json_file_path, 'w', encoding='utf-8') as jsonfile:
        json.dump(output_data, jsonfile, indent=4, ensure_ascii=False)
```

```python
convert_csv_to_custom_json("/content/drive/MyDrive/text2SQL/results/mistral_baseline_chinook_exp_res.csv", "/content/drive/MyDrive/text2SQL/results/Mistral_Baseline_Chinook_Prediction.json")
convert_csv_to_custom_json("/content/drive/MyDrive/text2SQL/results/mistral_rag_chinook_exp_res.csv", "/content/drive/MyDrive/text2SQL/results/Mistral_Rag_Chinook_Prediction.json")
```
