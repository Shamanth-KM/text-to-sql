````python
import pandas as pd
import re

# Function to apply the regex to each row
def apply_regex_to_column(value, regex_pattern):
    # Use re.search() or re.findall() depending on your need
    match = re.findall(regex_pattern, str(value), re.DOTALL)
    if match:
        try:
            matches = match[0][1].strip()
            sql_code = matches.replace('\n', ' ').replace('\t', ' ').replace('\\n', ' ').replace('```sql', ' ').replace('###', ' ')
            sql_code = re.sub(r'\s+', ' ', sql_code)
            return sql_code.strip()
        except:
            return None

    return value  # Return None if no match found
````

````python
# Load the CSV file
file_path = "/content/drive/MyDrive/text2SQL/results/mistral_baseline_spider_exp_res.csv"  # Update with your CSV file path
df = pd.read_csv(file_path)

# Specify the column you want to apply regex to and the column for results
input_column = 'clean_sql_extracted'  # Replace with your column name
output_column = 'clean_sql_extracted'  # Replace with your desired output column name

# Define your regex pattern (modify as needed)
regex_pattern = r"SQL:(.*?)\n*(.*?)(;|$)"  # Replace with your regex pattern

# Apply the regex function to the specified column and store the result in another column
df[output_column] = df[input_column].apply(lambda x: apply_regex_to_column(x, regex_pattern))

# Save the updated DataFrame back to a new CSV file
df.to_csv("/content/drive/MyDrive/text2SQL/results/mistral_baseline_spider_exp_res.csv", index=False)

#print("Data extraction and regex application complete. Output saved in 'output_file.csv'.")
````

````python
##Convert Prediction file to json for Birdbench
import csv
import json

def convert_csv_to_custom_json(csv_file_path, json_file_path):
    output_data = {}

    with open(csv_file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for idx, row in enumerate(reader):
            clean_sql = row.get("clean_sql_extracted", "").strip()
            db_id = row.get("db_id", "").strip()
            combined = f"{clean_sql}\t----- bird -----\t{db_id}"
            output_data[str(idx)] = combined

    with open(json_file_path, 'w', encoding='utf-8') as jsonfile:
        json.dump(output_data, jsonfile, indent=4, ensure_ascii=False)
````

````python
convert_csv_to_custom_json("/content/drive/MyDrive/text2SQL/results/mistral_baseline_spider_exp_res.csv", "/content/drive/MyDrive/text2SQL/results/Mistral_Baseline_Spider_Prediction.json")
convert_csv_to_custom_json("/content/drive/MyDrive/text2SQL/results/mistral_rag_spider_exp_res.csv", "/content/drive/MyDrive/text2SQL/results/Mistral_Rag_Spider_Prediction.json")
````
